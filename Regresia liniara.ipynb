## Exemplu regresie liniara

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel1 = np.poly1d(np.polyfit(x, y, 1))
mymodel2 = np.poly1d(np.polyfit(x, y, 2))
mymodel3 = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

fig = plt.figure()
ax = fig.add_subplot()
ax.set_title("Grad 1")
plt.scatter(x, y)
plt.plot(myline, mymodel1(myline))
plt.show()

fig = plt.figure()
ax = fig.add_subplot()
ax.set_title("Grad 2")
plt.scatter(x, y)
plt.plot(myline, mymodel2(myline))
plt.show()

fig = plt.figure()
ax = fig.add_subplot()
ax.set_title("Grad 3")
plt.scatter(x, y)
plt.plot(myline, mymodel3(myline))
plt.show()
```
